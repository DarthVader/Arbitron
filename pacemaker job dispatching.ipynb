```python
from collections import deque
import numpy as np
import pandas as pd
from settings import Settings
from database import Database
from pprint import pprint
```

```python
db = Database()
db.get_exchanges()
db.get_pairs()
df_exchanges = db.df_exchanges
df_pairs = db.df_pairs
```

```python
# calculating cycles
df_exchanges['cycles'] = np.floor(df_exchanges.ratelimit.max() / df_exchanges.ratelimit)
cycles = df_exchanges.set_index('id')[['cycles']].to_dict()['cycles']
pprint(cycles)
```

```python
# creating dict out of exchange-pair dataframe
pairs = pd.pivot_table(df_pairs[['exchange', 'pair']], index=['exchange'], aggfunc=deque).to_dict()['pair']
pairs
```

```python
jobs = []
for i in range(5):
    print("==========")
    print(" JOB", i+1)
    print("==========")
    for ex in df_exchanges.id.values:
        cycle = int(cycles[ex])
        job = list(pairs[ex])[:cycle]
        print(f"{ex}: {job}")
        pairs[ex].rotate(-cycle)
```
