```python
class Database():
    
    cassandra_nodes = ['127.0.0.1', '10.7.0.11', '10.7.0.20', '10.7.0.19']
    
    def __init__(self):    
        print("Connecting to database...".format(), end="", flush=False)
        try:
            self.cluster = Cluster(contact_points=self.cassandra_nodes, port=9042)
            self.session = self.cluster.connect(keyspace='arbitron')

        except Exception as e:
            print("{}".format(e))
        print("SUCCESS.") 
    
    
    def execute(self, cql):
        try:
            res = self.session.execute(cql, timeout=5)
            return pd.DataFrame(res._current_rows)
        
        except Exception as e:
            print("{}".format(e))
            return pd.DataFrame()
```

```python
db = Database()
```

```python
df = db.execute("select * from arbitron.history per partition limit 1")
```
