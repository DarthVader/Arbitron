```python
import altair as alt
alt.renderers.enable('notebook')
from vega_datasets import data
```

```python
cars = data.cars()
```

```python
cars.head(10)
```

```python
chart = alt.Chart(cars).mark_point().encode(
    x="Horsepower",
    y="Miles_per_Gallon",
    color="Acceleration"
)
chart.config = {'view': {'height': 400, 'width': 800}}
chart.interactive()
```
